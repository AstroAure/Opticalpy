```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
%matplotlib ipympl
import opticalpy as op
```

```python
scene = op.Scene(xlim=(-5, 50), ylim=(-30,30))

mirror = op.elems.FlatMirror(aperture=10, scene=scene, origin=[40,0], rotation=70)
ray = op.elems.CollimatedBeam(aperture=5, N_sources=1, scene=scene, origin=[0,0], rotation=0)

fig, ax = plt.subplots(figsize=(8,8))

axrot = fig.add_axes([0.25, 0.15, 0.65, 0.03])
rot = Slider(axrot, 'Rotation', 0.0, 180.0, 90.0)

def update(val):
    mirror.update_pos(new_pos=[40,0], new_rot=val)
    ax.cla()
    scene.plot(ax=ax, show=False)
rot.on_changed(update)

scene.plot(ax=ax, show=False)
plt.show()
```
